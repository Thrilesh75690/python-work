```python
# 1. Accept project marks, internal marks, external marks from student
# 2. passmarks is 50 in each subject
# 3. calculate total score if student got more than or equal to 50 in each subject.

# total_score= 70% on project + 20% on external =10% on internal
# print A grade if score is 90+
#       b grade if score b\w 70-90
#       c grade if score b\w 50-70
project_marks=int(input("enter project marks"))         #90
internal_marks=int(input("enter internal marks"))       #78
external_marks=int(input("enter external marks"))       #45
if project_marks>=50 and internal_marks>=50 and external_marks>=50:
  total=(project_marks*70/100)+(internal_marks*20/100)+(external_marks*10/10)
  print("total score is",total)
  if total>=90:
    print("grade is A")
elif total>=70:
   print("grade is B")
elif total>=50:
   print("grade is C")
```
